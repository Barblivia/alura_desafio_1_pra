##**1. Análisis de facturación**

```python
# Importación librerías

import matplotlib.pyplot as plt

# Cálculo de la facturación total de cada tienda
facturacion_tienda_1 = tienda1["Precio"].sum()/1000000
facturacion_tienda_2 = tienda2["Precio"].sum()/1000000
facturacion_tienda_3 = tienda3["Precio"].sum()/1000000
facturacion_tienda_4 = tienda4["Precio"].sum()/1000000

# Creación de un DataFrame con la facturación de cada tienda
dfFacturacion = pd.DataFrame({
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Facturacion Total": [facturacion_tienda_1, facturacion_tienda_2, facturacion_tienda_3, facturacion_tienda_4]
    })

# Mostrar el resultado
dfFacturacion["Facturacion Total"] = dfFacturacion["Facturacion Total"].apply(lambda x: "{:,.2f}".format(x))
print(dfFacturacion)

# Cambiar el formato a float para el gráfico
dfFacturacion["Facturacion Total"] = dfFacturacion["Facturacion Total"].str.replace(",", "").astype(float)

# Gráfico de barras

colores = ["#FF5733", "#6C91BF", "#2EC4B6", "#F1C40F"]
plt.bar(dfFacturacion["Tienda"], dfFacturacion["Facturacion Total"],color=colores)
plt.xlabel("Tiendas")
plt.ylabel("Facturación Total")
plt.title("Facturación por Tienda - En millones de pesos")
plt.show()

# Gráfico de torta con porcentajes de facturacion
plt.pie(dfFacturacion["Facturacion Total"],
        labels=dfFacturacion["Tienda"],
        autopct="%1.1f%%",
        colors=colores,
        startangle=90)
plt.title("Porcentaje de Facturación por Tienda")
plt.axis("equal")
plt.show()
```

##**2. Ventas por categoría**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Union de tablas y unificación de los colores
todas_categorias = pd.concat([tienda1, tienda2, tienda3, tienda4])["Categoría del Producto"].unique()
colores = plt.cm.tab20.colors  # paleta de 20 colores distintos
color_map = {cat: colores[i % len(colores)] for i, cat in enumerate(sorted(todas_categorias))}

# Función para gráfico por tienda en cantidades

def ranking_categorias(df, nombre_tienda):
    ranking = (df.groupby("Categoría del Producto")["Producto"]
               .count()
               .reset_index()
               .rename(columns={"Producto": "Cantidad Vendida"})
               .sort_values(by="Cantidad Vendida", ascending=False))

    # Colores según mapa global
    colores_categorias = [color_map[cat] for cat in ranking["Categoría del Producto"]]

    # Gráfico horizontal en cantidades
    fig, ax = plt.subplots()
    ax.barh(ranking["Categoría del Producto"], ranking["Cantidad Vendida"], color=colores_categorias)
    ax.set_xlabel("Cantidad Vendida")
    ax.set_ylabel("Categoría del Producto")
    ax.set_title(f"Categorías más vendidas - {nombre_tienda}")
    ax.invert_yaxis()

    # Etiquetas con valores
    for i, valor in enumerate(ranking["Cantidad Vendida"]):
        ax.text(valor + 0.5, i, str(valor), va='center', fontsize=9)

    plt.tight_layout()
    plt.show()


# Gráficos para las 4 tiendas
ranking_categorias(tienda1, "Tienda 1")
ranking_categorias(tienda2, "Tienda 2")
ranking_categorias(tienda3, "Tienda 3")
ranking_categorias(tienda4, "Tienda 4")


# Gráfico comparativo de productos entre tiendas

# Unificar las tiendas
tienda1["Tienda"] = "Tienda 1"
tienda2["Tienda"] = "Tienda 2"
tienda3["Tienda"] = "Tienda 3"
tienda4["Tienda"] = "Tienda 4"

df_total = pd.concat([tienda1, tienda2, tienda3, tienda4])

# Agrupar por categoría y tienda
ranking_total = (df_total.groupby(["Categoría del Producto", "Tienda"])["Producto"]
                 .count()
                 .reset_index()
                 .rename(columns={"Producto": "Cantidad Vendida"}))

# Orden global de categorías
orden_categorias = (ranking_total.groupby("Categoría del Producto")["Cantidad Vendida"]
                    .sum()
                    .sort_values(ascending=False)
                    .index)

# Colores por tienda
colores_tiendas = {
    "Tienda 1": "#FF5733",
    "Tienda 2": "#6C91BF",
    "Tienda 3": "#2EC4B6",
    "Tienda 4": "#F1C40F"
}

# Configuración  gráfico
fig, ax = plt.subplots(figsize=(10, 8))

categorias = list(orden_categorias)
num_tiendas = ranking_total["Tienda"].nunique()
bar_height = 0.18
y_positions = range(len(categorias))
offsets = [(-1.5 + i) * bar_height for i in range(num_tiendas)]

# Graficar barras agrupadas por tienda
for idx, tienda in enumerate(colores_tiendas.keys()):
    datos_tienda = ranking_total[ranking_total["Tienda"] == tienda].set_index("Categoría del Producto").reindex(categorias)
    cantidades = datos_tienda["Cantidad Vendida"].fillna(0)

    ax.barh(
        [y + offsets[idx] for y in y_positions],
        cantidades,
        height=bar_height,
        label=tienda,
        color=colores_tiendas[tienda]
    )

    # Números en el gráfico
    for y, valor in zip([y + offsets[idx] for y in y_positions], cantidades):
        if valor > 0:
            ax.text(valor + 0.5, y, str(int(valor)), va='center', fontsize=8)

# Ajustes grafico
ax.set_yticks(y_positions)
ax.set_yticklabels(categorias)
ax.invert_yaxis()
ax.set_xlabel("Cantidad Vendida")
ax.set_title("Categorías más vendidas por tienda (Comparativo)")
ax.legend(title="Tienda")

plt.tight_layout()
plt.show()
```

##**3. Calificación promedio de la tienda**

```python
import matplotlib.pyplot as plt
import pandas as pd

# Colores de las tiendas
colores_tiendas = {
    "Tienda 1": "#FF5733",
    "Tienda 2": "#6C91BF",
    "Tienda 3": "#2EC4B6",
    "Tienda 4": "#F1C40F"
}

# Colores para calificaciones
colores_calificaciones = {
    1: "#d73027",  # rojo intenso
    2: "#fc8d59",  # naranja claro
    3: "#fee08b",  # amarillo
    4: "#d9ef8b",  # verde claro
    5: "#1a9850"   # verde intenso
}

# DataFrames de las tiendas
tiendas = {
    "Tienda 1": tienda1,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

# Calcular promedios
df_promedios = pd.DataFrame({
    "Tienda": list(tiendas.keys()),
    "Promedio Calificación": [df["Calificación"].mean() for df in tiendas.values()]
}).sort_values(by="Promedio Calificación", ascending=False)

# Crear figura
fig = plt.figure(figsize=(14, 10))

# Promedio calificciones
ax1 = plt.subplot2grid((3, 2), (0, 0), colspan=2)
barras = ax1.bar(df_promedios["Tienda"], df_promedios["Promedio Calificación"],
                 color=[colores_tiendas[t] for t in df_promedios["Tienda"]])

ax1.set_title("Promedio de Calificación por Tienda", fontsize=14, fontweight="bold")
ax1.set_ylabel("Promedio de Calificación")
ax1.set_ylim(0, 5)

# Etiquetas de promedio en cada barra
for barra, val in zip(barras, df_promedios["Promedio Calificación"]):
    ax1.text(barra.get_x() + barra.get_width()/2, val + 0.05,
             f"{val:.2f}", ha="center", fontsize=11, fontweight="bold")

# Distribución de calificaciones
for i, (nombre_tienda, df) in enumerate(tiendas.items(), start=1):
    distribucion = df["Calificación"].value_counts().sort_index()
    colores_barras = [colores_calificaciones[cal] for cal in distribucion.index]
    total_clientes = distribucion.sum()

    ax = plt.subplot2grid((3, 2), divmod(i+1, 2), rowspan=1)
    wedges, texts, autotexts = ax.pie(distribucion.values,
                                      labels=distribucion.index,
                                      autopct='%1.1f%%',
                                      colors=colores_barras,
                                      startangle=90,
                                      textprops={'color':"black"})
    ax.set_title(f"{nombre_tienda} (Total: {total_clientes})",
                 color=colores_tiendas[nombre_tienda],
                 fontsize=12, fontweight="bold")

plt.tight_layout()
plt.show()
```

##**4. Productos más y menos vendidos**

```python
import matplotlib.pyplot as plt
import pandas as pd

# Paletas
colores_mas = ["#1a9850", "#66bd63", "#a6d96a"]  # verdes
colores_menos = ["#d73027", "#f46d43", "#fdae61"]  # rojos

# Layout 2x2 para las tiendas
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for idx, (nombre_tienda, df) in enumerate(tiendas.items()):
    # Agrupar facturación por producto
    facturacion = df.groupby("Producto")["Precio"].sum().sort_values(ascending=False)

    # Top 3 más y menos
    top3 = facturacion.head(3)
    bottom3 = facturacion.tail(3)

    # Unir para el gráfico (manteniendo orden)
    facturacion_seleccionada = pd.concat([top3, bottom3])
    colores = colores_mas[:len(top3)] + colores_menos[:len(bottom3)]

    # Graficar barras horizontales
    axes[idx].barh(facturacion_seleccionada.index[::-1],
                   facturacion_seleccionada.values[::-1] / 1_000_000,
                   color=colores[::-1])

    # Títulos
    axes[idx].set_title(f"{nombre_tienda} - Top y Bottom 3 por Facturación",
                        color=colores_tiendas[nombre_tienda], fontsize=12, fontweight="bold")
    axes[idx].set_xlabel("Facturación (Millones)")

    # Etiquetas al final de cada barra
    for i, val in enumerate(facturacion_seleccionada.values[::-1]):
        axes[idx].text(val / 1_000_000 + 0.02,
                       i,
                       f"{val/1_000_000:.2f}M",
                       va="center",
                       fontsize=9,
                       fontweight="bold")

plt.tight_layout()
plt.show()
```

##**5. Envío promedio por tienda**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Colores de las tiendas
colores_tiendas = {
    "Tienda 1": "#FF5733",
    "Tienda 2": "#6C91BF",
    "Tienda 3": "#2EC4B6",
    "Tienda 4": "#F1C40F"
}

# Comparativa de métricas
metricas_envio = []

for nombre_tienda, df in tiendas.items():
    promedio = df["Costo de envío"].mean()
    mediana = df["Costo de envío"].median()
    maximo = df["Costo de envío"].max()
    minimo = df["Costo de envío"].min()

    metricas_envio.append({
        "Tienda": nombre_tienda,
        "Promedio": promedio,
        "Mediana": mediana,
        "Máximo": maximo,
        "Mínimo": minimo
    })

tabla_envio = pd.DataFrame(metricas_envio).round(2)
print(" Métricas de Costo de Envío por Tienda")
print(tabla_envio)

# Porcentaje de envío sobre precio
metricas_porcentaje = []

for nombre_tienda, df in tiendas.items():
    df = df.copy()
    df["% Envío"] = (df["Costo de envío"] / df["Precio"]) * 100
    promedio_pct = df["% Envío"].mean()
    metricas_porcentaje.append({
        "Tienda": nombre_tienda,
        "Promedio % Envío": promedio_pct
    })

tabla_porcentaje = pd.DataFrame(metricas_porcentaje).round(2)
print("\n Promedio % de Envío sobre Precio por Tienda")
print(tabla_porcentaje)

#  Gráfico comparativo
plt.figure(figsize=(8, 6))
plt.bar(tabla_porcentaje["Tienda"],
        tabla_porcentaje["Promedio % Envío"],
        color=[colores_tiendas[t] for t in tabla_porcentaje["Tienda"]])

for i, val in enumerate(tabla_porcentaje["Promedio % Envío"]):
    plt.text(i, val + 0.2, f"{val:.2f}%", ha='center', fontsize=10)

plt.title("Promedio % de Costo de Envío sobre Precio", fontsize=14, fontweight='bold')
plt.ylabel("% de Envío sobre Precio")
plt.ylim(0, max(tabla_porcentaje["Promedio % Envío"]) + 2)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
